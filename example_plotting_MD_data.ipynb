```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('test.xlsx')

df.head()
```

```python
sns.set(style="ticks")

plot1 = sns.jointplot(df["Distance"], df["Angle"],kind='hex', 
              stat_func=None, xlim=(0, 10), ylim=(60,200),size=10);

plot1.ax_joint.plot([1.41],[175],'go')
plot1.x=df["Distance.1"]
plot1.y=df["Angle.1"]
# plot1.plot_joint(plt.scatter, marker='o', c='r', s=10)
```

```python
sns.set(style="ticks")

plot2=sns.jointplot(df["Distance.1"], df["Angle.1"],kind='hex', color='r',
              stat_func=None, xlim=(0, 10), ylim=(60,200),size=10,
              );

plot2.ax_joint.plot([1.41],[175],'go')
```
